```python
#sistema de grids
#parametros
cor_fundo = '#f5f5f5'
grid = sns.FacetGrid(db, col='Territorialidades', hue='Territorialidades', col_wrap=4)

#adicionando graficos de linhas em cada grafico do sns
grid = grid.map( plt.plot, 'Ano', 'PIB per capita')

#adicionando sombra + ajuste do titulo
grid = grid.map(plt.fill_between, 'Ano', 'PIB per capita', alpha=0.2).set_titles('{col_name} Territorialidades')
#alterando o titulo para o nome da coluna
grid = grid.set_titles('{col_name}')

#adicionando o subtitulo
grid = grid.fig.suptitle(
    'Evolução da Renda per capita por estado\n by @Lucas Gomes do Nascimento'
)
plt.subplots_adjust(top=0.93)
```
